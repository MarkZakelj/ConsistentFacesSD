```python
import numpy as np
import pandas as pd
```

```python
import plotly.express as px
from utils.metrics import construct_dataframe_faces
from utils.imgs import get_bbox_size
import plotly.graph_objects as go
```

```python
subsets = [
    'base_one_person_dreamshaper',
    'base_two_people_dreamshaper',
    'ip_one_person',
    'ip_one_person_wp5',
    'ip_one_person_wp25',
    'two_people_faceid_dreamshaper',
]
subset_name = subsets[0]
```

```python
df_raw_1 = construct_dataframe_faces(subsets[2])
df_raw_2 = construct_dataframe_faces(subsets[3])
df_raw_3 = construct_dataframe_faces(subsets[4])
dfs = [df_raw_1, df_raw_2, df_raw_3]
```

```python
for i, df in enumerate(dfs):
    df['bbox_size'] = df['bbox'].apply(lambda x: get_bbox_size(x))
    dfs[i] = df.sort_values('bbox_size', ascending=False).groupby('image_id').first().reset_index()
```

```python
def add_source(dframe, source_name, columns_to_keep):
    dframe = dframe[columns_to_keep].copy()
    dframe['Source'] = source_name
    return dframe
```

```python
df_main = pd.concat([
    add_source(dfs[0], 'ip_one_person', ['face_quality']),
    add_source(dfs[1], 'ip_one_person_wp5', ['face_quality']),
    add_source(dfs[2], 'ip_one_person_wp25', ['face_quality']),
])
```

```python
df_main
```

```python
fig = px.box(df_main, x='Source', y='face_quality', title='Face quality comparison when using different weights on IP adapters')
fig.update_layout(
    width=1000,  # Set a custom width in pixels
    height=600,
    autosize=True, 
    plot_bgcolor='white',
    paper_bgcolor='white', 
    font=dict(color='black'),  # Set font color to black for better contrast
)
```
